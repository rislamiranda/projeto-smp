##Descrevendo os dados do ANO 2019

```python
def analise_descritiva(df):
    #tipo de dado que tenho em cada coluna do dataframe (ou quantidade, ou qualidade)

    # CNPJ (quantidade)
    print(f"CNPJ - ", len(df['CNPJ'].unique()))

    print()
    # Codigo nacional (quantidade)
    print(f"Código Nacional - ", len(df['Código Nacional'].unique()))

    print()
    # Município (quantidade)
    print(f"Município - ", len(df['Município'].unique()))

    print()
    # UF (dados)
    print(f"UF - ", len(df['Município'].unique()))
    print(f"UF - ", df['UF'].unique())

    print()
    # Modalidade de Cobrança (quantidade)
    print(f"Modalidade de Cobrança - ", len(df['Modalidade de Cobrança'].unique()))
    # Modalidade de Cobrança (dados)
    print(f"Modalidade de Cobrança - ", df['Modalidade de Cobrança'].unique())

    print()
    # Tecnologia (quantidade)
    print(f"Tecnologia - ", len(df['Tecnologia'].unique()))
    # Tecnologia (dados)
    print(f"Tecnologia - ", df['Tecnologia'].unique())

    print()
    # Tecnologia Geração (quantidade)
    print(f"Tecnologia Geração - ", len(df['Tecnologia Geração'].unique()))
    # Tecnologia Geração (dados)
    print(f"Tecnologia Geração - ", df['Tecnologia Geração'].unique())

    print()
    # Empresa (quantidade)
    print(f"Empresa - ", len(df['Empresa'].unique()))
    # Empresa (dados)
    print(f"Empresa - ", df['Empresa'].unique())

    print()
    # Porte da Prestadora (quantidade)
    print(f"Porte da Prestadora - ", len(df['Porte da Prestadora'].unique()))
    # Porte da Prestadora (dados)
    print(f"Porte da Prestadora - ", df['Porte da Prestadora'].unique())

    print()
    # Tipo de Pessoa (quantidade)
    print(f"Tipo de Pessoa - ", len(df['Tipo de Pessoa'].unique()))
    # Tipo de Pessoa (dados)
    print(f"Tipo de Pessoa - ", df['Tipo de Pessoa'].unique())

    print()
    # Tipo de Produto (quantidade)
    print(f"Tipo de Produto - ", len(df['Tipo de Produto'].unique()))
    # Tipo de Produto (dados)
    print(f"Tipo de Produto - ", df['Tipo de Produto'].unique())

    print()
    # Código IBGE Município (quantidade)
    print(f"Código IBGE Município - ", len(df['Código IBGE Município'].unique()))

    print()
    # Grupo Econômico (quantidade)
    print(f"Grupo Econômico - ", len(df['Grupo Econômico'].unique()))
    # Grupo Econômico (dados)
    print(f"Grupo Econômico - ", df['Grupo Econômico'].unique())


```

```python
analise_descritiva(df_2019)
```

```python
def NaoNumero_1(df):
    
    # vendo os Not a Number (NaN) do 1º semestre e sua relevância face ao total de informações do dataset

    print("1º semestre de 2019:")

    print(f"Quantidade de NaN: ", df_2019_1[['2019-01', '2019-02', '2019-03', '2019-04',
       '2019-05', '2019-06']].isna().sum().sum())

    # O shape do 1º semestre é
    print(f"Linhas e colunas: ", df_2019_1.shape)

    # numero total de entradas no meu dataframe
    print(f"Total de entradas no meu df: ", df_2019_1.shape[0] * df_2019_1.shape[1])

    # vendo a percentagem no Nan na tabela do 1º semestre face ao total de informações
    dado1 = df_2019_1[['2019-01', '2019-02', '2019-03', '2019-04',
       '2019-05', '2019-06']].isna().sum().sum() / (df_2019_1.shape[0] * df_2019_1.shape[1])

    print(f"No 1º semestre de 2019, os dados NaN representam {dado1*100:.2f}% do total de entradas no dataframe")
```

```python
def NaoNumero_2(df):
    # vendo os Not a Number (NaN) do 2º semestre e sua relevância face ao total de informações do dataset

    print("2º semestre de 2019:")

    print(f"Quantidade de NaN: ", df_2019_2[['2019-07', '2019-08', '2019-09', '2019-10',
       '2019-11', '2019-12']].isna().sum().sum())

    # O shape do 1º semestre é
    print(f"Linhas e colunas: ", df_2019_2.shape)

    # numero total de entradas no meu dataframe
    print(f"Total de entradas no meu df: ", df_2019_2.shape[0] * df_2019_2.shape[1])

    # vendo a percentagem no Nan na tabela do 2º semestre face ao total de informações
    dado2 = df_2019_2[['2019-07', '2019-08', '2019-09', '2019-10',
       '2019-11', '2019-12']].isna().sum().sum() / (df_2019_2.shape[0] * df_2019_2.shape[1])

    print(f"No 2º semestre de 2019, os dados NaN representam {dado2*100:.2f}% do total de entradas no dataframe")
    #vendo os NaN da tabela com 'merge + outer'

    print("Ano de 2019:")

    print(f"Quantidade de NaN: ", df_2019[['2019-01', '2019-02', '2019-03', '2019-04',
       '2019-05', '2019-06', '2019-07', '2019-08', '2019-09', '2019-10',
       '2019-11', '2019-12']].isna().sum().sum())

    # O shape do ano é
    print(f"Linhas e colunas: ", df_2019.shape)

    # numero total de entradas no meu dataframe
    print(f"Total de entradas no meu df: ", df_2019.shape[0] * df_2019.shape[1])

    # vendo a percentagem no Nan na tabela 2019 face ao total de informações
    dado_t = df_2019[['2019-01', '2019-02', '2019-03', '2019-04',
       '2019-05', '2019-06', '2019-07', '2019-08', '2019-09', '2019-10',
       '2019-11', '2019-12']].isna().sum().sum() / (df_2019.shape[0] * df_2019.shape[1])

    print(f"No ano de 2019, os dados NaN representam {dado_t*100:.2f}% do total de entradas no dataframe")

    print()
    print(f"Apesar de a tabela consolidada apresentar {dado_t*100:.2f}% de entradas NaN, essas entradas são \
irrelevantes, pois: \n a) tabela do 1º semestre tem apenas {dado1*100:.2f}% de NaN; \n b) a tabela do 2º semestre \
tem apenas {dado2*100:.2f}% de NaN; \n c) com o merge das tabelas do 1º e 2º semestre, todos os campos sem \
foram preenchidos com 'NaN', e \n são esses NaN preenchidos automaticamente pelo sistema que \
estão na tabela consolidada")
```
