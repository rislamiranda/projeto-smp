Segundo o Guia de Análise de Atos de Concentração Horizontal do Cade (“Guia H”), o “Índice Herfindahl-Hirschman
(HHI) pode ser utilizado para o cálculo do grau de concentração dos mercados. O HHI é calculado com base no
somatório do quadrado das participações de mercado de todas as empresas de um dado mercado. O HHI pode chegar
até 10.000 pontos, valor no qual há um monopólio, ou seja, em que uma única empresa possua 100% do mercado.”

Exemplo de cálculo para o Estado do Acre:

(0.14**2) + (8.34**2) + (0.73**2) + (0.37**2) + (40,61**2) + (12.55**2) + (0.16**2) + (37.10**2)

Calculando a contribuição de cada grupo economico para o calculo do HHI da região

para o cálculo do HHI, primeiro criamos uma coluna chamada hhi_empresa pegando o valor do market_share 
de cada empresa e elevando o número ao quadrado, e multiplicamos este valor por 10_000
O calculo do HHI leva em consideração a % de market share na base 100, ou seja: 40%, 30% etc
como definimos que para o Power BI era melhor deixar o market share em 4 casas decimais e sem a %
temos que multiplicar o resultado da potencia por 10_000, pois:
(40**2) = 1_600 ---> ((0.40*10^2)**2) = 1_600

```python
df_2021_uf['hhi_empresa'] = (df_2021_uf['market_share'] ** 2) * 10_000

# ver se vamos chamar de hhi_empresa mesmo, ou se mudamos para hhi_grupo
# também pode ser mudado após a consolidação de todas as tabelas
```

```python
# Segundo o Guia de Análise de Atos de Concentração Horizontal do Cade (“Guia H”), o “Índice Herfindahl-Hirschman
# (HHI) pode ser utilizado para o cálculo do grau de concentração dos mercados. O HHI é calculado com base no
# somatório do quadrado das participações de mercado de todas as empresas de um dado mercado. O HHI pode chegar
# até 10.000 pontos, valor no qual há um monopólio, ou seja, em que uma única empresa possua 100% do mercado.”
```

```python
# Calculando a contribuição de cada grupo economico para o calculo do HHI da região

# para o cálculo do HHI, primeiro criamos uma coluna chamada hhi_empresa pegando o valor do market_share 
# de cada empresa e elevando o número ao quadrado, e multiplicamos este valor por 10_000
# O calculo do HHI leva em consideração a % de market share na base 100, ou seja: 40%, 30% etc
# como definimos que para o Power BI era melhor deixar o market share em 4 casas decimais e sem a %
# temos que multiplicar o resultado da potencia por 10_000, pois:
# (40**2) = 1_600 ---> ((0.40*10^2)**2) = 1_600

df_2021_ddd['hhi_empresa'] = (df_2021_ddd['market_share'] ** 2) * 10_000

# ver se vamos chamar de hhi_empresa mesmo, ou se mudamos para hhi_grupo
# também pode ser mudado após a consolidação de todas as tabelas
```
