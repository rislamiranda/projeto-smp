```python
# nas colunas dos meses, cerca de metade dos dados são nan
# # mas isso nao desqualifica a análise, pois:
# a) os dados da Anatel estão organizados por semestre (1 tabela para jan a jun e 1 tabela 
# para jul a dez)
# b) ao conferir a tabela original para ver como esses NaN constavam, nas células respectivas
# não constava nenhuma informação: a célula estava em branco (não eram strings e nem caracteres
# especiais, por exemplo)
# c) assim, na tabela original, essas células estavam zeradas: ou nao haviam chips ativos naquele 
# mês, por linha de dados (empresa, cliente, tecnologia etc.), ou a empresa não preencheu os dados
# d) mesclamos as duas tabelas de cada ano por meio da operação merge, parâmetro auter
# e) ao fazer o merge, vimos que a tabela que, antes tinha 19 colunas (13 colunas com dados majoritariamente 
# de strings e 6 colunas de dados de acesso, referente aos meses jan-jun), passou a ter 25 colunas, 
# pois recebeu a 6 colunas dos meses jul-dez da 2ª tabela 
# e) para as linhas que tinham exatamente as mesmas informações nas 13 colunas precedentes de strings (CNPJ,                                                                                                
# Empresa, Porte da Pessoa, tipo de Produto, Tecnologia etc.), as informações de ambos os semestres foram preenchidas
# f) mas, em casos em que não havia identidade de grafia nas informações das 13 colunas, o python jogou os 
# dados relativos ao 2º semestre para uma linha completamente nova, preenchendo o espaço referente aos 
# meses jan-jul com NaN
# g) nessa migração, a tabela consolidada passou a contar com uma quantidade bastante grande de Nan
# h) passou-se, então, para a análise de como tratar esses dados
```
