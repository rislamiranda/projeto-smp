```python
#criar uma função para checar a quantidade de NaN presentes nos dataframes
def check_nas(lista_df):
    lista_nas = []
    for df in lista_df:
        valor_na = df.isna().sum().sum()
        lista_nas.append(valor_na)
    lista_qnt_x_total = []
    for qnt_na in lista_nas:
        qnt_x_total = round((qnt_na * 100) / quant_células_valor, 2)
        lista_qnt_x_total.append(qnt_x_total)
    return lista_qnt_x_total
```

```python
#criar uma função para verificar a concentração de mercado a partir do HHI
def analise_hhi(valor_hhi):
    if valor_hhi <= 1500:
        return "pouco concentrado"
    if valor_hhi > 1500 and valor_hhi <= 2500:
        return "moderado"
    else:
        return "altamente concentrado"
```
