```python
# chamando as bibliotecas

import os
import glob

# os é a biblioteca para interagir com o sistema operacional
# glob é um módulo da biblioteca padrão do Python, que torna a correspondência de caminhos de arquivos e 
# a realização de diversas tarefas relacionadas a arquivos uma tarefa mais simples e eficaz
```

```python
# criando uma funcao que vai popular uma lista com os nomes dos arquivos que terminam com a extensão
# csv e estao no diretorio especificado

# preencher a variável 'diretorio' com o caminho para a pasta que contém os arquivos:

def lista_arquivos_csv(diretorio = './data/tecnologia'):
    return(glob.glob(os.path.join(diretorio, '*.csv')))

# criamos uma função para listar todos os arquivos CSV contidos em um diretório específico. 
# def lista_arquivos_csv(diretorio): estamos definindo o nome da nossa função, e ela aceira como argumento
# um diretorio (para chamar essa funçãotemos que fornecer um diretório como argumento
# os.path.join(diretorio, '*.csv'): esta parte da função cria um caminho completo para os arquivos CSV 
# no diretório especificado
# os.path.join() é usado para combinar o caminho do diretório (diretorio) com o padrão de arquivo '*.csv', 
# o que resulta em algo como diretorio/*.csv.
# glob.glob(...): a função glob.glob() é usada para encontrar todos os arquivos que correspondam ao 
# padrão especificado no diretório fornecido: ele procura todos os arquivos que terminam com ".csv" 
# no diretório especificado e popula uma lista com esses arquivos    
```

```python
lista_arquivos_csv()
```

```python
def ano_arquivo(nome_arquivo):
    parte = (nome_arquivo.split('_')[-2])
    return parte

# criando uma função que vai percorrer os meus arquivos e separar (split) os elementos do nome do 
# arquivo quando encontrar um padrão que eu defini (no caso, um underline '_'), começando, de traz
# pra frente[-], e fazendo o split no segundo [-2] travessão
```

```python
import pandas as pd 
from pandas import DataFrame


def le_baseRS(prefixo,sufixo):
    df_consolidado = DataFrame()
    for arquivo in lista_arquivos_csv():
        if arquivo.find(prefixo) != -1 and arquivo.find(sufixo) != -1:      # para verificar que tem o prefixo e sufixo    
            ano = ano_arquivo(arquivo)
            df_ano = pd.read_csv(arquivo, sep=';', header=0)
            df_ano['ano']=ano
            df_consolidado = pd.concat([df_consolidado,df_ano])
    return df_consolidado


df_tec_tipo = le_baseRS('tec','tipo')
df_tec_geracao = le_baseRS('tec','geracao')
df_tec_modalidade = le_baseRS('tec','modalidade')
df_tec_5g_por_estado = le_baseRS('tec','5g_por_estado')
df_tec_5g_por_municipio = le_baseRS('tec','5g_por_municipio')
```
